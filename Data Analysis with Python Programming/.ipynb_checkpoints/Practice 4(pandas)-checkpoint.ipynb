```python
import pandas as pd
import numpy as np
```

```python
# Pandas Series.
# Creating Pandas Series.
# Becareful with the braces and the brackets that you uses. Python interprets them differently.
g7_pop = pd.Series([35.467, 63.951, 80.940, 60.665, 127.061, 64.511, 318.523])
g7_pop.name = 'G7 Population in Millions'
g7_pop
```

```python
# Showing the datatype of the Series
type(g7_pop.values)
```

```python
# Indexing in a Series
g7_pop.index
```

```python
# We use normal indexing as in numpy arrays
g7_pop[6]
```

```python
# Labeling or changing the index defaul value.
g7_pop.index = [
    'Canada',
    'France',
    'Germany',
    'Italy',
    'Japan',
    'United Kingdom',
    'United States',
    ]
g7_pop
```

```python
g7_pop
```

```python
# Bypassing the default index values.
population = pd.Series({
    'Canada': 35.467,
    'France': 63.951,
    'Germany': 80.940,
    'Italy': 60.665,
    'Japan': 127.061,
    'Unite Kingdom': 64.511,
    'Unite States': 318.523,
}, name = 'Population in Millions')
population
```

```python
# Label Indexing
population['France']
```

```python
# Multi-indexing.
pd.Series(population , index = ['France', 'Japan', 'Germany'])
```

```python
#Sequencial Indexing
g7_pop.iloc[5]
```

```python
#Creating a Pandas Series.
geochemicalInfo = pd.Series({
    'X': 408391,
    'Y': 9768300,
    'Ti': 0.698,
    'V': 0.354,
    'Cr': 0.241,
    'Mn': 2.447,
    'Fe': 88.197,
    'Co': 0.889,
    'Ni': 0.845,
    'Cu': 2.826,
    'Zn': 0,
    'Zr': 0,
    'Nb': 0,
    'Mo': 1.295,
    'Ag': 0,
    'Sn': 0,
    'Sb': 0,
    'W': 0.661,
    'Pb': 0,
    }, name = 'Geochemical Information of a paricular location.')
geochemicalInfo
```

```python
# Indexing using label
geochemicalInfo['Zn']
```

```python
# Indexing using the iloc function
geochemicalInfo.iloc[7]
```

```python
geochemicalInfo.index
```

```python
population.index
```

```python
# Boolean Series.
geochemicalInfo < 1
```

```python
# Showing the values in a boolean series.
geochemicalInfo[geochemicalInfo < 1]
```

```python
# Using 'and' in a boolean series
geochemicalInfo[(geochemicalInfo < 1) & (geochemicalInfo > 0.2)]
```

```python
# Using 'or' in a boolean series
geochemicalInfo[(geochemicalInfo < 0.5) | (geochemicalInfo > geochemicalInfo.mean())]
```

```python
# Using 'not' in a boolean series
geochemicalInfo[~(geochemicalInfo < 0.2) & (geochemicalInfo > 1)]
```
