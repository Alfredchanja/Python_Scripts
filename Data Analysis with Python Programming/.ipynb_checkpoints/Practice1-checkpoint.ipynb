```python
import numpy as np
```

```python
# Creating an array.
np.array([1, 2, 3, 4, 5])
```

```python
# Accessing the elements of an array.
a = np.array([1, 2, 3, 4, 5])
a[0], a[1]
```

```python
#Indexing and slicing an array.
a[0:]
```

```python
a[0:3]
```

```python
a[:4]
```

```python
a[2:]
```

```python
a[2:4]
```

```python
a[[0, 2, -1]]# This is called multi-indexing. This creats another array.
```

```python
# Array types.
a.dtype
```

```python
# Changing the array type.
np.array([1, 2, 3, 4, 5], dtype=np.float)
```

```python
# Dimesions and shapes.
# This is a 2D array.
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

A.shape
```

```python
A.ndim
```

```python
A.size
```

```python
# A 3D array
B = np.array([[
            [1, 2, 3],
            [4, 5, 6],
            [10, 11, 12],
        ],
        [
            [2, 3, 4],
            [1, 5, 2],
            [34, 25, 12],
        ]
])
B.shape
```

```python
B.ndim
```

```python
B.size
```

```python
# Indexing and slicing multiple dimension arrays.
B[1]
```

```python
B[1, 0, 2]
```

```python
B[0, :2]
```

```python
B[1, 1:, 2:]
```

```python
# Modifying an array.
B[1, 0] = np.array([34, 52, 89])
```

```python
print(B)
```

```python
B[0, 1, 2] = 22
print(B)
```

```python
# Summary Statistics
B.sum()
```

```python
B.mean()
```

```python
B.std()
```
