```python
import numpy as np
```

```python
# Numpy Array Creation
# Creates an array filled with zeros.
a = np.zeros(10)
print(a)
```

```python
# Creates a 2*2 array filles with 1s
b = np.ones([2, 2], dtype=np.int8)
print(b)
```

```python
#Creates a 3*2 array filles with 4s.
c = np.ones([3, 2], dtype=np.int8) * 4
print(c)
```

```python
# Creates an array with random integer values/uninitialized values.
d = np.empty(5, dtype=np.int8)
print(d)
```

```python
# Creates an array of a constant value of specific data type.
e = np.full([3, 3], 12, dtype=np.int8)
print(e)
```

```python
# Creating an array that is evenly spaced.
f = np.arange(10)
print(f)
```

```python
# Creating an evenly spaced array ranging from 0 to 20.
g = np.arange(0, 20)
print(g)
```

```python
# Creating an evenly space array ranging from 0 to 1 with a spacing of 0.1
h = np.arange(0, 1, 0.1)
print(h)
```

```python
# Returning a sequence evenly spaced value in the range of 0 to 1
j = np.linspace(10, 20, 20)
print(j)
```

```python
# Creating random arrays with random values.
k = np.random.random ([2, 2])
print(k)
```

```python
l = np.random.normal([2, 2])
print(l)
```

```python
m = np.random.rand(2,4)
print(m)
```

```python
o = np.arange(8).reshape(2, 4)
print(o)
```

```python
p = np.arange(8).reshape(2, 2, 2)
print(p)
```
